```python
# Team now wants to target different types of spenders
# Create spending flag for each user based on criteria
# 1. if the mean of prices of products purchased by a user is lower than 10
#    then flag them as a "low spender"
# 2. if the mean of prices of products purchased by a user is higher than or equal to 10
#    then flag them as a "High spender"

# split the data into groups based on user_id
# apply the transform() on prices to generate average price by user
# create a new column, mean_price to place results
ords_prods_merge['mean_price'] = ords_prods_merge.groupby('user_id')['prices'].transform(np.mean)
```

```python
ords_prods_merge[['user_id', 'mean_price']].head(1)
```

```python
# use results and derive new column spending_flag

ords_prods_merge.loc[ords_prods_merge['mean_price'] < 10, 'spending_flag'] = 'Low spender'

ords_prods_merge.loc[ords_prods_merge['mean_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# QA check flag
ords_prods_merge[['user_id','mean_price', 'spending_flag']].head(100)
```

```python
# Team wants to distinguish frequent versus non-frequent customers to send relevant notifications if they want to buy the same item again
# Create order frequency flag that makrs the regularity of a user's ordering behavior
    # according to the median in the "days_since_prior_order" column
# Criteria
# 1. if the median of “days_since_prior_order” is higher than 20, 
    #then the customer should be labeled a “Non-frequent customer.”
# 2. If the median is higher than 10 and lower than or equal to 20, 
    # then the customer should be labeled a “Regular customer.”
# 3. If the median is lower than or equal to 10, 
    # then the customer should be labeled a “Frequent customer.”
    
# split the data into groups based on user_id
# apply the transform() on days prior to generate median 
# create a new column, median_days_since_prior_order to place results
ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby('user_id')['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge[['user_id', 'days_since_prior_order','median_days_since_prior_order']].head(11)
```

```python
# Use results and derive new column order_frequency_flag

ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'

ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] <= 20) & (ords_prods_merge['median_days_since_prior_order'] > 10), 'order_frequency_flag'] = 'Regular customer'

ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] <= 10, 'order_frequency_flag'] = 'Frequent customer'

```
